```r
dc<-read.csv("data//dc-wikia-data.csv")
marvel<-read.csv("data//marvel-wikia-data.csv")
par(cex.axis=0.6,font.axis=2)
plot(dc$ALIGN,col="blue")
title("ALign of DC")
plot(marvel$ALIGN,col="red")
title("ALign of Marvel")
```

```r
par(cex.axis=0.6,font.axis=2)
plot(marvel$ALIVE,marvel$SEX)
title("Alive characters by sex from DC")
plot(dc$ALIVE,dc$SEX)
title("Alive characters by sex from DC")
```

```r
sex1 <- sort(table(dc$HAIR),
            decreasing = T)

sex2 <- sort(table(marvel$HAIR),
            decreasing = T)

barplot(sex1, las = 2, cex.names = 0.7)
title("Bar chart for DC")
barplot(sex2, las = 2, cex.names = 0.7)
title("Bar chart for Marvel")
```

```r
sex1 <- sort(table(dc$ALIGN),
            decreasing = T)
pie(sex1)
title("Pie chart for DC")
sex1 <- sort(table(marvel$ALIGN),
            decreasing = T)
pie(sex1)
title("Pie chart for Marvel")
```
